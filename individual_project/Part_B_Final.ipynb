# Topic: Number of Passengers Carried through AirTransport compared to labour force participation and population who visited for studies between age 20-24.

- Source of data & graphs: THE WORLD DEVELOPMENT EXPLORER
- Timeline: 2010 - 2019

```python
df = pd.read_csv('https://raw.githubusercontent.com/kq49539/DATA690-ATMAN/main/individual_project_part_B_draft/wdi_data%20(1).csv')
df.head(10)
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
df
```

```python
df_country = df[df["Country Code"].isin(['USA','CHN','JPN','GBR','DEU','IRL','BRA','IND','TUR','RUS'])]
df_country
```

```python
df_air = df_country.query("indicator == 'IS.AIR.PSGR'")
df_air
```

# Bar Graph
- Indicators: Country Name vs Air Transport Passenger carried

```python
fig = go.Figure()

trace_0=go.Bar(
    x = df_air["Country Name"],
    y = df_air["value"]
)

fig.add_trace(trace_0)

fig.update_layout(
    title="Top 10 Countries by 2019 Air Transport, Passengers Carried",
    xaxis_title="Country Name",
    yaxis_title="2010 Air Transport, Passengers Carried"
)

fig.show()
```

# Chart Interpretation

- The Bar graph depicts that the number of passengers carried from 2010 to 2019 to the United States was more than 900M. Whereas, China was the 2 country which carried maximum number of passengers. 
- It was followed by United Kingdom, Ireland,India and others respectively.

# Time Series

- Indicators: Number of passengers carried vs year

```python
fig =go.Figure()

trace_0=go.Line(
    x = df_air["Year"],
    y = df_air["value"]
)

fig.update_layout(
    title = "Top 10 Countries by 2019 Air Transport, Passengers Carried",
    xaxis={"title":"Year"},
    yaxis={"title":"Air Transport Passengers Carried"}
)

fig.add_trace(trace_0)
fig.show()
```

```python
df_pop = df_country.query("indicator == 'SP.DYN.LE00.IN'")
df_pop
```

# Scatter Plot

- Indicators: Barro-Lee population of age 20-24 vs AirTransport passengers carried 

```python
fig =go.Figure()

trace_0=go.Scatter(
    x = df_air["value"],
    y = df_pop["value"],
     mode = "markers"
)

fig.update_layout(
    title = "Top 10 Countries by 2019 Air Transport, Passengers Carried",
    xaxis={"title":"2010 Air Transport, Passenger Carried"},
    yaxis={"title":"Barro-Lee: Population, age 20-24"}
    
)

fig.add_trace(trace_0)
fig.show()
```

# Chart Interpretation

- Number of passengers whose age was between 20-24 were carried through AirTransport.

```python
df_labour = df_country.query("indicator == 'SL.TLF.ACTI.1524.ZS'")
df_labour
```

```python
fig =go.Figure()

trace_0=go.Scatter(
    x = df_air["value"],
    y = df_labour["value"],
     mode = "markers"
)

fig.update_layout(
    title = "Top 10 Countries by 2019 Air Transport, Passengers Carried",
    xaxis={"title":"2010 Air Transport, Passenger Carried"},
    yaxis={"title":"2010 Labour Participation rate for ages 15-24"}
    
)

fig.add_trace(trace_0)
fig.show()
```

# Chart Interpretation

- The figure elicits that the number of passengers for Brazil and UK were more involved in labour force.
